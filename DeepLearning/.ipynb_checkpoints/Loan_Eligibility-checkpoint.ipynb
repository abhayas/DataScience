```python
plt.figure()
sns.countplot(data=sortedbysubgrade, x="sub_grade", hue="loan_status")
plt.title('Default-Sized Countplotr')
plt.figure(figsize=(12,6))
plt.show()
```

```python
subgradefng = ["F1","F2","F3","F4","F5","G1","G2","G3","G4","G5"]
```

```python
plt.figure()
sns.countplot(data=sortedbysubgrade[sortedbysubgrade["sub_grade"].isin(subgradefng).sort_values()], x="sub_grade", hue="loan_status")
plt.title('Default-Sized Countplotr')
plt.figure(figsize=(12,6))
plt.show()
```
